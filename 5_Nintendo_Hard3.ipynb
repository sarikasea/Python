What percentage of records have missing values in at least one column? Handle the missing values, so that we have a cleaned dataset to work with.

```python
# Calculate the percentage of rows with missing values
missing_rows_percentage = (df.isnull().any(axis=1).sum() / len(df)) * 100
print(f"Percentage of records with missing values: {missing_rows_percentage:.2f}%")

# Handle missing values
# Fill missing 'region' and 'demographic_group' with the mode
df['region'] = df['region'].fillna(df['region'].mode()[0])
df['demographic_group'] = df['demographic_group'].fillna(df['demographic_group'].mode()[0])

# Fill missing 'pre_order_quantity' with the mean
df['pre_order_quantity'] = df['pre_order_quantity'].fillna(df['pre_order_quantity'].mean())

# Display the cleaned DataFrame
display(df)
```

```python
# Convert 'pre_order_date' to datetime objects
df['pre_order_date'] = pd.to_datetime(df['pre_order_date'])

# Extract the month and year from the date
df['order_month'] = df['pre_order_date'].dt.to_period('M')

# Group by region, demographic group, and month and sum the pre_order_quantity
monthly_orders = df.groupby(['region', 'demographic_group', 'order_month'])['pre_order_quantity'].sum().reset_index()

# Display the results
display(monthly_orders)
```

Predict the total pre-sales quantity for each region for September 2024. Assume that growth rate from August to September, is the same as the growth rate from July to August in each region.

```python
# Calculate total pre-orders per region for each month
regional_monthly_orders = monthly_orders.groupby(['region', 'order_month'])['pre_order_quantity'].sum().reset_index()

# Pivot the table to have months as columns
regional_monthly_pivot = regional_monthly_orders.pivot(index='region', columns='order_month', values='pre_order_quantity').reset_index()

# Get the period objects for July and August
july_period = pd.Period('2024-07', freq='M')
august_period = pd.Period('2024-08', freq='M')

# Calculate growth rate from July to August
regional_monthly_pivot['growth_rate_jul_aug'] = (regional_monthly_pivot[august_period] - regional_monthly_pivot[july_period]) / regional_monthly_pivot[july_period]

# Predict September orders
regional_monthly_pivot['predicted_sept_2024'] = regional_monthly_pivot[august_period] * (1 + regional_monthly_pivot['growth_rate_jul_aug'])

# Display the predicted September orders
display(regional_monthly_pivot[['region', 'predicted_sept_2024']])
```
