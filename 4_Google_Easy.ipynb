```python
from google.auth import default
creds, _ = default()
gc = gspread.authorize(creds)
```

```python
sheet = gc.open('Python Summer Challenge').worksheet('Q4-1')
user_engagement_data = get_as_dataframe(sheet)
df = user_engagement_data
df
```

Identify and remove any duplicate entries in the dataset to ensure data quality. How many duplicates were found and removed?

```python
initial_row_count = df.shape[0]
df_cleaned = df.drop_duplicates()
duplicates_removed_count = initial_row_count - df_cleaned.shape[0]

print(f"Number of duplicate entries found and removed: {duplicates_removed_count}")
df = df_cleaned
```

After dropping duplicates, aggregate the data to find the average user interaction time for each number of search results displayed per page. What are the average interaction times?

```python
average_interaction_time = df.groupby('search_results_displayed')['interaction_time'].mean()
print("Average interaction time for each number of search results displayed:")
display(average_interaction_time)
```

Sort the aggregated results from Q2 to determine which number of search results per page has the highest average user interaction time. What is the optimal number of search results per page?

```python
sorted_average_interaction_time = average_interaction_time.sort_values(ascending=False)
print("Average interaction time sorted in descending order:")
display(sorted_average_interaction_time)
```
